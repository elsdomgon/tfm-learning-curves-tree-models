```python
import pandas as pd
```

## Visualización de resultados

```python
# BÚSQUEDA ALEATORIA
print(f"{' BÚSQUEDA ALEATORIA ':=^50}")

# CSV principal
df_principal_random = pd.read_csv('resultados_LSTM_efectoT_random.csv')
display(df_principal_random)

# CSV por folds
df_folds_random = pd.read_csv('resultados_LSTM_folds_efectoT_random.csv')
display(df_folds_random)


# ALGORITMO TPE
print(f"\n{' ALGORITMO TPE ':=^50}")

# CSV principal
df_principal_TPE = pd.read_csv('resultados_LSTM_efectoT_TPE.csv')
display(df_principal_TPE)

# CSV por folds
df_folds_TPE = pd.read_csv('resultados_LSTM_folds_efectoT_TPE.csv')
display(df_folds_TPE)
```

## Parámetros e importancia de los mismos

```python
# Lista de valores de T 
valores_T = [5, 10, 15, 20]


# BÚSQUEDA ALEATORIA
print(f"{' RESULTADOS: BÚSQUEDA ALEATORIA ':=^60}\n")

for i in range(4):
    print(f"[T = {valores_T[i]}]")
    print(f"Importancia parámetros: {df_principal_random['Importancia_Params'][i]}")
    print(f"Configuración: {df_principal_random['Parámetros'][i]}\n")


# ALGORITMO TPE
print(f"\n{' RESULTADOS: ALGORITMO TPE ':=^60}\n")

for i in range(4):
    print(f"[T = {valores_T[i]}]")
    print(f"Importancia parámetros: {df_principal_TPE['Importancia_Params'][i]}")
    print(f"Configuración: {df_principal_TPE['Parámetros'][i]}\n")
```
